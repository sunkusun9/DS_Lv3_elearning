### 2. 표본 추출(Sampling)
- 표본 추출(Sampling): 모집단에서 표본을 선택하는 과정

#### 비확률 표본 추출법
- 각 추출 대상이 표본으로 선정될 확률을 계산하지 않는 추출법
- 모집단을 정확하게 규정지을 수 없는 경우
- 표본오차가 큰 문제가 되지 않는 경우
- 비용, 시간, 조사의 편리함 때문에 자주 사용

| 방법 | 설명 |
| --- | --- |
| 간편 추출법 | 자의적인 판단에 따라 추출 |
| 판단 추출법 | 조사자의 나름의 지식과 경험에 의해 모집단을 잘 대표한다고 판단되는 표본을 추출 |
| 할당 추출법 | 조사목적과 밀접하게 관련된 변수에 따라 부분집단으로 구분하고 구성 비율에 따라 추출 |
| 눈덩이 추출법 | 접근이 어렵거나 추출의 방향을 작성하기 곤란한 특정 집단에 대한 조사 |

#### 분포의 모양

- 왜도(Skewness): 분포의 치우침

$g_1 = \frac{\frac{1}{N} \sum_{i=1}^{N} (x_i - \mu)^3}{\left( \frac{1}{N} \sum_{i=1}^{N} (x_i - \mu)^2 \right)^{3/2}}$

| 왜도 | 모양 | 특징 |
| --- | --- | --- |
| 양의 왜도 | 우측으로 긴 꼬리 | 평균>중앙값>최빈값 |
| 0 | 칭 | 평균=중앙값=최빈값 |
| 음의 왜도 | 좌측으로 긴 꼬리 | 평균<중앙값<최빈값 |

slide13-image4


- 첨도(Kurtosis): 응집도

$\gamma_2 = \frac{\frac{1}{N} \sum_{i=1}^{N} (x_i - \mu)^4}{\left( \frac{1}{N} \sum_{i=1}^{N} (x_i - \mu)^2 \right)^2}$

- 표준 첨도(Excess kurtosis)

$g_2 = \gamma_2 - 3$

> 첨 첨도: 뾰족함, 꼬리가 두꺼움
> 중간 첨도: 정규분포와 동일한 뾰족함
> 완만 첨도: 평평함, 꼬리 얇음

### 4. 확률의 개념

#### 용어
- 시행(Trial, Experiment): 결과가 우연에 의해 정해지는 행위 또는 실험
- 결과(Outcome): 시행을 통해 실제로 관찰된 단일 결과
- 표본공간(Sample Space, S or Ω): 전체 결과의 목록
- 사건(Event): 표본 공간의 부분 집합으로, 관심의 대상

#### 수학적 확률
- 이론적으로 계산한 확률
- 논리와 이론적 모델에 의해 계산

모든 경우가 발생할 가능성이 동일하다면,
$P(A)=\frac{A사건의 경우의 수}{전체 경우의 수}$

#### 통계적 확률
- 실험이나 관찰을 통해 얻은 경험적 확률

$P(A)=\frac{A 사건의 관측수}{전체 관측수}$
- 실험을 반복함에 따라 신뢰도 향상

#### 확률의 기본 성질

|성질|내용|
| --- | --- |
| 확률의 범위 | 어떤 사건 A 의 확률, $0≤𝑃(𝐴)≤1$|
| 표본 공간 전체확률 | 표본 공간 S 의 확률, $𝑃(𝑆)=1$|
| 불가능한 사건의 확률 | 일어날 수 없는 사건의 확률, $𝑃(∅)=0$|
| 여사건 확률 | 어떤 사건 A 가 일어나지 않을 확률 $𝑃(𝐴)^𝑐=1−𝑃(𝐴)$ |
| 합사건 | 사건 A 또는 B 가 일어날 확률, <br/>$𝑃(𝐴\cup𝐵)=𝑃(𝐴)+𝑃(𝐵)−𝑃(𝐴\cap𝐵)$<br/>사건 A 와 B 가 겹치지 않을 때(배반 사건, $𝑃(𝐴\cap𝐵)=0)$<br/>$𝑃(𝐴∪𝐵)=𝑃(𝐴)+𝑃(𝐵)$|

#### 결합 확률(확률의 곱셈)

$𝑃(𝐴\cap𝐵)=𝑃(𝐴|𝐵)𝑃(𝐵)=𝑃(𝐵|𝐴)𝑃(𝐴)$

사건 A, B 가 서로 독립 이면,

$𝑃(𝐴|𝐵)=𝑃(𝐴), 𝑃(𝐵|𝐴)=𝑃(𝐵)$

$𝑃(𝐴\cap𝐵)=𝑃(𝐴)𝑃(𝐵)$



#### 조건부 독립

$𝑃(𝐴\cap𝐵|𝐶)=𝑃(𝐴|𝐶)𝑃(𝐵|𝐶)$
$𝑃(𝐴|𝐵,𝐶)=𝑃(𝐴|𝐶)$

- 조건부 독립의 예
𝐷: 질병의 유무, 𝑆: 질병의 증상의 유무, 𝑇: 질병의 검사 결과

$𝑃(𝑆\cap𝑇)\ne𝑃(𝑆)𝑃(𝑇)$

$D → S, D → T ▷ 𝑃(𝑆|𝑇,𝐷)=𝑃(𝑆|𝐷)$


#### 전확률의 법칙(Law of Total Probability)

- 사건 의 표본공간이 배타적인 k 개의 공간으로 나누었을 때,

$\forall i \neq j,\; B_i \cap B_j = \emptyset, \qquad
S = \bigcup_{i=1}^{k} B_i
$

#### 베이즈 정리(Bayes’s Theorem)

- $𝑃(𝐵|𝐴)=\frac{𝑃(𝐴|𝐵)}{𝑃(𝐴)}𝑃(𝐵)$
> $𝑃(𝐴\cap𝐵)=𝑃(𝐴|𝐵)𝑃(𝐵)$
> 
> $𝑃(𝐴\cap𝐵)=𝑃(𝐵|𝐴)𝑃(𝐴)$
> 
> $𝑃(𝐵|𝐴)𝑃(𝐴)=𝑃(𝐴|𝐵)𝑃(𝐵)$

- B의 표본공간이 서로 배타적인, k 개의 $B_i$공간으로 나누어졌을 때,

S = \bigcup_{i=1}^{k} B_i
