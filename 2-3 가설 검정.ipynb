#### p-value 계산

$𝒑−𝒗𝒂𝒍𝒖𝒆=𝑃_𝜃(𝑅𝑒𝑗𝑒𝑐𝑡 𝐻_0|𝐻_0 𝑖𝑠 𝑇𝑟𝑢𝑒)$

소비자 고발원에 OO 치즈의 슬라이스 당 무게가 20g이 되지 않는다는 사건이 접수가 되었습니다.

포장 기계에는 발생할 수 표준편차가 0.4g 이라고 적혀 있었고, 이 기계에서 발생하는 오차는 맞다고 가정합니다.

이를 위해 30개의 표본 슬라이스의 무게를 측정한 결과를 가지고 (30개는 표본의 평균이 정규분포를 따르기에 충분한 표본 수라고 가정합니다.)

소비자의 주장을 받아들일 수 있는지 분석을 해봅니다. (유의 수준: 0.05)

치즈 슬라이스 표본의 무게(g)

```
19.37, 20.20, 19.91, 19.20, 19.57, 20.46, 18.83, 19.63, 20.31, 19.45,
```

````python
# 수집된 증거
X = [
    19.37, 20.20, 19.91, 19.20, 19.57, 20.46, 18.83, 19.63, 20.31, 19.45, 
    19.53, 19.76, 20.40, 19.54, 19.62, 19.63, 20.68, 20.67, 20.20, 19.95, 
    20.09, 20.40, 19.43, 20.27, 19.30, 19.54, 20.16, 19.23, 19.74, 19.46
]
````

**[풀이]**

<div style="border: 1px solid #ddd; padding: 12px; margin-top: 10px">

**pvalue 계산에 활용될 수 있는 정보**

- $H_0$ 가 맞다는 가정하에서 나오는 정보: μ(모평균) = 20g


- 문제에서 나오는 정보: σ(모표준편차) = 0.4g


- 표본수 30개는 표본평균의 분포가 정규분포를 따른다고 할 수 있습니다.


**증거의 확률 분포 및 통계량**


- 표본의 평균 $\sim N(20, \frac{0.4^2}{30})$ 


- 검정 통계량(Z) = $\frac{\bar{X}-20}{0.4/\sqrt{30}}$


**검정의 종류**

- **왼쪽(left) 꼬리 검정**:검정 통계량보다 작은 값이 대립 가설을 더 지지합니다. 
    
</div>

````python
from scipy.stats import norm, t
````

````python
# 표준정규 분포를 이용한 pvalue 도출을 그래프로 설명
X_bar = np.mean(X)

mu_X_bar = 20
sigma_X_bar = 0.4 / (30 ** 0.5)

# 검정통계량
Z = (X_bar - mu_X_bar) / sigma_X_bar

plt.figure(figsize=(6, 4))
plt.plot(np.linspace(-5, 5, 100), norm.pdf(np.linspace(-5, 5, 100)))
plt.axvline(Z, color='red', linestyle='--')
plt.text(Z, 0.01, str(Z)[:6])
plt.fill_between(np.linspace(-5, Z, 100), np.zeros(shape=100),
                norm.pdf(np.linspace(-5, Z, 100)), color='orange')
plt.show()
````

### 3. 모평균 검정: 단일 모집단

$𝑯_𝟎:𝝁=𝝁_𝟎$
- 모분산($\sigma^2$)을 알고 있을 경우
> 모집단이 정규 분포 를 따르거나, 중심극한정리(Central Limit Theorem)를 따를 만큼 표본수가 충분할 때

$표본 평균 \sim N(\mu_0, \frac{\sigma^2}{n})$

$𝒁=\frac{\bar{𝑿}−𝝁_𝟎}{\frac{𝝈}{\sqrt{𝒏}}}$


- 모분산($\sigma^2$)를 모를 경우 
> 모집단이 정규 분포 를 따른다면,

$t=\frac{\bar{𝑿}−𝝁_𝟎}{\frac{S}{\sqrt{𝒏}}}$

$t \sim t(n-1)$

### 4. 모평균 검정: 두 개의 모집단

$𝑯_𝟎:𝝁_1=𝝁_2$
- 모분산($\sigma_1^2$, $\sigma_2^2$)을 알고 있을 경우
> 두 표본이 독립이고, 모집단이 정규분포를 따른다면, 

$𝒁=\frac{\bar{X}_1−\bar{X_2}}{\sqrt{\frac{\sigma_1^2}{n_1}+\frac{\sigma_2^2}{n_2}}}$


- 모분산($\sigma_1^2, \sigma_2^2$)를 모를 경우 
> 두 표본이 독립이고, 모집단이 정규 분포 를 따른다면,

||두 모집단이 등분산|두 모집단이 이분산|
|---|----|----|
|검정 통계량|$\frac{\bar{X_1} - \bar{X_2}}{S_p\sqrt{1/n_1 + 1/n_2}}$<br/>$S_p^2 = \frac{(n_1 - 1)S_1^2 + (n_2 - 1)S_2^2}{n_1 + n_2 - 2}$|$\frac{\bar{X_1} - \bar{X_2}}{\sqrt{S_1^2/n_1 + S_2^2/n_2}} \sim t(\nu)$|
|자유도|$n_1+n_2-2$|$\frac{\left(S_1^2/n_1 + S_2^2/n_2\right)^2}{\left(\frac{S_1^2}{n_1}\right)^2/(n_1 - 1) + \left(\frac{S_2^2}{n_2}\right)^2/(n_2 - 1)}$|

### 5. 모평균 검정: 대응 관계인 두 개의 모집단
$𝑯_𝟎:𝝁_𝟏=𝝁_𝟏′$
- 검정 통계량
$𝐷=𝑋_1−𝑋_1′$
D 의 모집단이 정규 분포 를 따른다면,

$𝑡=\frac{\bar{𝐷}}{\frac{𝑆_𝑑}{\sqrt{𝑛}}} \sim 𝑡(𝑛−1)$


### 6.모비율 검정: 단일 모집단
$𝑯_𝟎:𝒑=𝒑_𝟎$

$\hat{𝑝}=\frac{𝑥}{𝑛}$, 𝑥: Positive(Success)의 수, n: 표본수

표본수가 충분하여, 𝑝 ̂가 중심극한정리에 의해 정규분포를 따른다고 할 수 있을 때
- 검정통계량

$𝑍=\frac{𝑝−𝑝_0}{\sqrt{\frac{𝑝_0(1−𝑝_0)}{𝑛}}}$



### 7.모분산 검정: 단일 모집단
$𝑯_𝟎:𝝈^𝟐= 𝝈_𝟎^𝟐$
- 모집단이 정규 분포 를 따를 경우

검정 통계량: $\chi^2=(𝑛−1)\frac{𝑆^2}{𝜎_0^2}~\chi^2(𝑛−1)$


### 8. 모분산 검정: 두 개의 모집단

- 두 모집단이 정규 분포 를 따를 경우
 
검정통계량: $𝐹=\frac{𝑆_1^2}{𝑆_2^2}~𝐹(𝑑1,𝑑2)$

$𝑑_1=𝑛_1−1, 𝑑_2=𝑛_2−1$
