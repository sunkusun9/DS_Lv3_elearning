### 1. 확률 분포
#### 확률 변수(Random Variable)
- 확률 실험의 결과를 수치로 나타낸 변수
- 변수명은 보통 대문자, 소문자는 특정 관측치
```
Ex) 주사위 던지기
S = {1, 2, 3, 4, 5, 6}
X: 주사위 값이 홀수: 0, 짝수: 1
```

#### 기대값(Expected Value)

- 확률 변수의 평균적인 결과
- 무수히 많은 확률 변수를 추출하여 얻어지는 결과의 평균
- 이산 확률 변수

$E[X]=\sum_{\forall{𝑥}}x𝑓(𝑥)$

- 연속 확률 변수

$E[x]=\int_{-\infty}^{\infty} xf(x)\,dx$


#### 기대값의 성질
- 선형성

$𝐸[𝑎𝑋+𝑏]=𝑎𝐸[𝑋]+𝑏$
, a,b 는 임의의 상수
- 덧셈 법칙

$𝐸[𝑋+𝑌]=𝐸[𝑋]+𝐸[𝑌]$
- 곱셈 법칙

X, Y 두 확률변수가 독립 이라면 성립

$𝐸[𝑋𝑌]=𝐸[𝑋]𝐸[𝑌]$

#### 확률 변수의 분산(Variance)

- 기대값을 기준으로 값들이 얼마나 퍼져 있는지 나타내는 지표 
- '확률 변수값과 기대값의 편차 제곱'의 기대값

$𝑉𝑎𝑟(𝑋)=𝐸[(𝑋−𝜇)^2], 𝜇=𝐸[𝑋]$
- 이산 확률 변수

$Var(X)=\sum_{\forall x}(x−μ)^22f(x)$
- 연속 확률 변수

$Var(x)=\int_{-\infty}^{\infty} (x-\mu)^2f(x)\,dx$

▷ $𝑉𝑎𝑟(𝑋)=𝐸[𝑋^2]−𝐸[𝑋]^2$


#### 분산의 성질

$𝑉𝑎𝑟(𝑎𝑋+𝑏)=𝑎^2𝑉𝑎𝑟(𝑋)$ a,b 는 임의의 상수

$𝑉𝑎𝑟(𝑋+𝑌)=𝑉𝑎𝑟(𝑋)+𝑉𝑎𝑟(𝑌)+2𝐶𝑜𝑣(𝑋,𝑌)$

$𝐶𝑜𝑣(𝑋,𝑌)=𝐸[(𝑋−𝐸[𝑋])(𝑌−𝐸[𝑌])]=E[XY]-E[X]E[Y]$

▷ X, Y 두 확률변수가 독립 이라면, 

$𝐸[𝑋𝑌]=𝐸[𝑋]𝐸[𝑌]$ → $𝐶𝑜𝑣(𝑋,𝑌)=0$

$𝑉𝑎𝑟(𝑋+𝑌)=𝑉𝑎𝑟(𝑋)+𝑉𝑎𝑟(𝑌)$

### 5. 연속 확률 분포의 성질

#### 지수 분포의 무기억성

- 미래 사건의 확률은 이전 경과 시간의 영향을 받지 않음
   모든 s, t 에 대해
   $𝑃(𝑋>𝑠+𝑡|𝑋>𝑠)=𝑃(𝑋>𝑡)$

#### 감마 분포와 카이제곱 분포

- $𝜒^2(d)≡𝐺𝑎𝑚𝑚𝑎(\frac{𝑑}{2},\frac{1}{2})$

#### 카이제곱 분포의 성질

$s^2=\frac{1}{n-1}\sum_{i=1}^{n}\left(X_i - \bar{X}\right)^2$

$(n-1)s^2=\sum_{i=1}^{n}\left(X_i -\mu + \mu - \bar{X}\right)^2$

$(n-1)s^2=\sum_{i=1}^{n}\left(\left(X_i -\mu\right)^2 + 2 \left(X_i -\mu \right) \left(\mu - \bar{X}\right) + \left(\mu - \bar{X}\right)^2\right)$

> $\sum_{i=1}^{n}2 \left(X_i -\mu \right) \left(\mu - \bar{X}\right) = \sum_{i=1}^{n}2 \left(X_i \mu - X_i \bar{X} - \mu^2 + \mu\bar{X}\right)$
>
> $= 2n \left(\bar{X}\mu - \bar{X}^2 - \mu^2 + \mu\bar{X}\right)$
>
> $= -2n\left(\bar{X} - \mu \right)^2$

$(n-1)s^2=\sum_{i=1}^{n}\left(X_i -\mu\right)^2 - 2n\left(\bar{X} - \mu \right)^2 + n \left(\mu - \bar{X}\right)^2$

$     = \sum_{i=1}^{n}\left(X_i -\mu\right)^2- n\left(\bar{X} - \mu \right)^2$

$(n-1)\frac{s^2}{\sigma^2}=\sum_{i=1}^{n}\left(\frac{X_i -\mu}{\sigma}\right)^2 - n\left(\frac{\bar{X} - \mu}{\sigma}\right)^2=\sum_{i=1}^{n}\left(\frac{X_i -\mu}{\sigma}\right)^2 - \left(\frac{\bar{X} - \mu}{\frac{\sigma}{\sqrt{n}}}\right)^2$

> $\sum_{i=1}^{n}\left(\frac{X_i -\mu}{\sigma}\right)^2 \sim \chi^2_n, \left(\frac{\bar{X} - \mu}{\frac{\sigma}{\sqrt{n}}}\right)^2 \sim \chi^2_1$
>
> $\chi^2(n-1) = \chi^2(n) - \chi^2(1)$

따라서, $(n-1)\frac{s^2}{\sigma^2} \sim \chi^2(n-1)$


#### 표준 정규 분포와 t-분포

- 표준 정규 분포

> $X \sim N(\mu, \sigma^2)$
>
> $Z=\frac{X-\mu}{\sigma}$
>
> $Z \sim N(0, 1^2)$

- 표본 평균

> $\bar{X} = \frac{\sum_{i=1}^nX_i}{n}= \frac{(X_1 + X_2 + ... + X_n)}{n} $
>
> $E(\bar{X})= E(\frac{X_1 + X_2 + ... + X_n}{n}) = \frac{E(X_1) + E(X_2) + .. E(X_n)}{n} $
> 
>$X_1, ... X_n$의 모평균이 $\mu$이므로 $E(X_1), ... E(X_n)$ 모두 $\mu$
> 따라서, $E(\bar{X}) = \frac{n\mu}{n} = \mu$


>$V(\bar{X}) = V(\frac{X_1 + X_2 + ... + X_n}{n})$
>
> **n개의 표본이 서로 독립이라면,**
>
> $X_1, ... X_n$ 간의 공분산은 모두 $COV(X_1, X_2) = 0, ... COV(X_{n-1}, X_n) = 0$
>
> $V(\bar{X}) = V(\frac{X_1}{n}) + V(\frac{X_2}{n}) ... + V(\frac{X_n}{n})$
>
> $X_1, ... X_n$은 모두 분산이 $\sigma^2$이므로 $V(X_1), ... V(X_n)$ 모두 $\sigma^2$ 이고, $V(aX) = a^2V(X)$이므로
> $V(\bar{X}) = \frac{\sigma^2}{n^2} + \frac{\sigma^2}{n^2}... + \frac{\sigma^2}{n^2} = \frac{\sigma^2}{n}$

$\frac{\bar{X}-\mu}{\frac{\sigma}{\sqrt{n}}} \sim Z$

$\frac{\bar{X}-\mu}{\frac{s}{\sqrt{n}}} \sim t(n-1)$


$\frac{\bar{X} - \mu}{\frac{S}{\sqrt{n}}}=\frac{\bar{X} - \mu}{\frac{S}{\sqrt{n}}\frac{\sigma}{\sigma}}=\frac{\bar{X} - \mu}{\frac{\sigma}{\sqrt{n}}\frac{S}{\sigma}}=\frac{\bar{X} - \mu}{\frac{\sigma}{\sqrt{n}}\sqrt{\frac{(n-1)S^2}{(n-1)\sigma^2}}}=\frac{\bar{X} - \mu}{\frac{\sigma}{\sqrt{n}}\sqrt{\frac{(n-1)\frac{S^2}{\sigma^2}}{(n-1)}}}$

$=\frac{Z}{\sqrt{\frac{(n-1)\frac{s^2}{\sigma^2}}{n-1}}} \sim \frac{Z}{\sqrt{\frac{\chi^2(n-1)}{n-1}}} \sim t(\nu), \nu=n-1$

#### F-분포의 유례

- 두 모집단의 표본분산 / 모분산의 비율의 분포

$\frac{𝑆_1^2/𝜎_1^2}{𝑆_2^2/𝜎_2^2} \sim 𝐹(𝑛_1−1, 𝑛_2−2)$

$𝐹(𝑑_1,𝑑_2)=\frac{\chi^2(𝑑_1)/𝑑_1}{\chi^2(𝑑_2)/𝑑_2}$


$S_1^2$: 1번 모집단의 표본 분산,  $\sigma_1^2$: 1번 모집단의 모분산, $n_2$: 1번 모집단의 표본수

$S_2^2$: 2번 모집단의 표본 분산,  $\sigma_2^2$: 2번 모집단의 모분산, $n_2$: 2번 모집단의 표본수


- 유도 과정

$\frac{𝑆_1^2/𝜎_1^2}{𝑆_2^2/𝜎_2^2} = \frac{\frac{(n_1-1)S_1^2}{(n_1-1)\sigma_1^2}}{\frac{(n_2-1)S_2^2}{(n_2-1)\sigma_2^2}}=\frac{(n_1-1)\frac{S_1^2}{\sigma_1^2}\frac{1}{(n_1-1)}}{(n_1-1)\frac{S_2^2}{\sigma_2^2}\frac{1}{(n_2-1)}} \sim \frac{\frac{\chi^2(n_1-1)}{n_1-1}}{\frac{\chi^2(n_2-1)}{n_2-1}}=F(n_1-1, n_2-2)$

### 6. 적률

- 적률(Moment): 확률 분포의 모양 설명

$E[X^n] =
\begin{cases}
\displaystyle\sum_{i=1}^{n} x_i^{n} f(x_i), & \text{(이산형)} \\[10pt]
\displaystyle\int_{-\infty}^{\infty} x^{n} f(x)\,dx, & \text{(연속형)}
\end{cases}$

| 적률 | 수식 | 의미 |
| --- | --- | --- |
| 1차 적률 |$𝐸[𝑋]$| 평균(mean), 기대값 |
| 2차 적률 |$𝐸[𝑋^2]$| 분산 계산에 사용<br/> $Var(X)=E[X^2]-E[X]^2$|
| 3차 적률 |$𝐸[𝑋^3]$| 왜도 계산에 사용 |
| 4차 적률 |$𝐸[𝑋^4]$| 첨도 계산에 사용 |

#### 적률 생성 함수(Moment Generating Function)

- 확률 분포의 기대값, 분산, 왜도 등을 계산

| 분포 | 적률 생성 함수 |
| --- | --- |
| 베르누이 Bern(p) |$1−𝑝+𝑝𝑒^𝑡$|
| 이항 Bin(n,p) |$(1−𝑝+𝑝𝑒^𝑡)^n$|
| 포아송 Pois(λ) |$exp[\lambda(e^t-1)]$|
| 지수분포 Exp(λ) |$\frac{\lambda}{\lambda-t}$|
| 감마분포 Gamma(α, β) |$\left(\frac{\beta}{\beta-t}\right)^\alpha$|
| 정규분포 N(μ, σ 2) |$\exp[\mu t+\frac{1}{2}\sigma^2t^2]$|

- 적률 생성 함수가 없는 분포도 있음. Ex) 카우시(Cauchy) 분포

#### 적률 생성 함수(Moment Generating Function) 활용 예

- 포아송 분포

$M_X(t) = exp[\lambda(e^t-1)]$

$E[X] = M_X'(0), M_X'(t)= \lambda exp[\lambda(t+\lambda(e^t-1)]$

$E[X] = \lambda$


$E[X^2] = M_X''(0), M_X''(t)= \lambda(1+\lambda e^t)exp[\lambda(t+\lambda(e^t-1)]$

$E[X^2] = \lambda(1+\lambda)$

$Var(X) = E[X^2] - E[X]^2 = \lambda(1+\lambda) - \lambda^2 = \lambda$

#### 중심 극한 정리(Central Limit Theorem, CLT)

- 동일한 확률 분포를 가진 독립 확률 변수 n 개의 평균의 분포는 n 의 크기가 충분히 크다면 정규 분포를 따름
