### 2. 데이터이 유형

#### 정형 데이터(Structured Data)

- 정해진 형식과 구조
- 빠른 검색 및 집계 가능

Ex) DataFrame - 행 인덱스와 열 인덱스의 식별자를 지닌 2차원 자료 구조

| id | species | sepal length | sepal width | petal length | petal width |
| --- | --- | --- | --- | --- | --- |
| 1 | setosa | 5.1 | 3.5 | 1.4 | 0.2 |
| 2 | virginica | 4.9 | 3.0 | 1.4 | 0.2 |
| 3 | versicolor | 7.0 | 3.2 | 4.7 | 1.4 |


#### 반정형 데이터(Semi-structured data)

- 형식과 구조가 유연
- Key-Value 구조
- JSON, XML, HTML, …
- 데이터의 유형
```xml
[
    {
        "id": 1, 
        "species": "setosa",
        "size": {
            "sepal_length": 5.1, "sepal_with": 3.5,
            "petal_length": 1.4, "petal_width": 0.2
        }
    }, {
        "id": 2, 
        "species": "versicolor",
        "size": {
            "sepal_length": 6.0, "sepal_with": 2.2,
            "petal_length": 4.0, "petal_width": 1.0
        },
        "color": "blue"
    }
]
```


#### 비정형 데이터(Unstructured data)

- 데이터의 유형
- 구조가 정해지지 않은 데이터
- 분석을 위해 고급 처리 기술 필요
- 동영상, 이미지, 음성

### 3. 데이터의 종류

|   |정량적 데이터<br/>(수치자료, Quantitative Data) | 정성적 데이터<br/>(질적자료, Qualitative Data) |
| --- | --- | --- |
| 구성 | 숫자, 수치 | 텍스트, 이미지, 동영상, 등등 |
| 관점 | 객관적 | 주관적 |
| 분석 방법 | 통계분석, 모델링 | 코드화, 테마 추출, 사례 비교 |
| 특징 | “얼마나”를 설명에 유리 | “이유“의 설명에 유리 |


#### 정량적 데이터(Quantitative Data)의 예

| 포켓몬 | 체력 | 공격 | 방어 | 스피드 | 키(cm) | 몸무게(kg) |
| --- | --- | --- | --- | --- | --- | --- |
| 피카츄 | 35 | 55 | 40 | 90 | 40 | 6.0 |
| 리자몽 | 78 | 84 | 78 | 100 | 170 | 90.5 |
| 꼬부기 | 44 | 48 | 65 | 43 | 50 | 9.0 |

#### 정성적 데이터(Qualitative Data)의 예

| 포켓몬 | 외형 | 색상 | 표정 | 키워드 | 선호계층 |
| --- | --- | --- | --- | --- | --- |
| 피카츄 | 작고 노란 쥐형, <br/> 귀에 검은 무늬, 번개꼬리 | 노랑 | 귀엽고 밝음 😊 | 귀여움, 친근함, <br/>에너지틱 | 어린이, 여성층 |
| 리자몽 | 거대한 날개 달린 용, <br/> 화염을 내뿜는 꼬리 | 주황 | 위엄 있고 강인함 😠 | 용맹함, 카리스마, 불꽃 | 청소년, <br/> 게임 마니아|
| 꼬부기 | 파란 거북이, <br/> 갈색 등껍질, 동그란 눈 | 파랑 | 장난기 있고 친근함 😋 | 유쾌함, 귀여움, <br/> 단단함 | 유아층, <br/> 클래식 팬 |


### 4. 빅데이터란?
- 빅데이터
- 규모가 너무 커서 데이터 처리를 위한 고도화된 연산 체계가 필요한 데이터

#### 빅데이터 특징

| 광의 | 협의 | 정의 | 특징 |
| --- | --- | --- | --- |
| 5V | 3V | Volume(규모) | 데이터 양의 급격한 증가<br/>→ 전통적 데이터 시스템으로 처리에 한계 |
|   |   | Variety(유형) | 데이터 종류와 소스의 다양화<br/>→ 정형 데이터 뿐만 아니라 반정형, 비정형 데이터로 확대 |
|   |   | Velocity(속도) | 데이터가 생성되고 전송되는 속도가 매우 빠름<br/>→ 데이터 수집 및 처리의 고속화 |
|   | 2V | Veracity(품질) | 데이터의 신뢰성, 정확성, 타당성 보장이 필수 |
|   |   | Value(가치) | 대용량 데이터에서는 가치는 내재되어 있음 방대한 데이터에서 인사이트를 도출할 수 있는 역량이 필요 |


### 5. 데이터의 구성 요소

| 구성 요소 | 설명 | 유의어 | 실체 |
| --- | --- | --- | --- |
| 단위<br/>(Unit) | 분석의 대상 | 엔터티(Entity) | 관측치, 레코드, 인스턴스, 표본, 사례,<br/>데이터 포인트… |
| 변수<br/>(Variable) | 단위(Unit)의 특성 | 속성(Attribute)<br/>특성(Feature) | 변수값, 속성값, 특성값, … |


#### 예제: Iris 데이터

| --- | --- |
| 변수(Variable) | id |
|   | species |
|   | sepal length |
|   | petal length |
|   | petal width |


### 6. 분류 기준

| 분류 기준 | 구분 | 변수 |
| --- | --- | --- |
| 역할 | 분석 / 설명 / 예측에서<br/>대상이 되는 변수 | 출력 변수(Output variable)<br/>종속 변수(Dependent variable)<br/> 결과 변수(Outcome variable) |
|   | 대상을 변수를<br/>분석 / 설명 / 예측에<br/>활용하기 위한 변수 | 입력 변수(Input variable)<br/>독립 변수(Independent variable)<br/>예측 변수(Predictor variable) |


### 7. 변수의 구성 요소

| 구성 요소 | 실체 |
| --- | --- |
| 변수명 | 변수를 식별하기 위한 고유의 이름 Ex) sepal width, distance, gender,… |
| 측정 단위 | 어떤 단위로 나타낸 것인지 명시 Ex) ℃, ℓ, ㎞, … |
| 변수 설명 | 변수가 나타내는 실제 개념과 속성을 설명<br/>Ex) gender - 0: 남성 / 1: 여성<br/>sepal width - 붓꽃 꽃받침의 너비 |


#### 형식
| 형식 |   |   | 설명 |
| --- | --- | --- | --- |
| 범주형<br/>(Categorical) | 명목형(Nominal) |   | 대소/선후를 비교 할 수 없는 상징적인 값 |
|   | 순서형(ordinal) |   | 대소/선후를 비교가 가능한 상징적인 값 |
| 수치형<br/> (Numerical) | 가산성<br/>Countable | 이산형<br/>(Discrete) | 간격이 있고, 열거 가능하여 셀 수 있는 값 |
|   |   | 연속형<br/>(Continuous) | 간격의 개념이 없고, 열거가 불가능하여 셀 수 없는 값 |
|   | 0 의 의미<br/>Zero | 등간형(Interval) | 0 은 상대적 기준점 <br/> 동일한 수치상 차이가 동일한 정도의 차를 나타낸 값 <br/>비율의 개념으로 설명할 수 없는 값 |
|   |   | 비율형<br/>(Ratio) | 0 의 개념이 절대적 없음을 의미<br/>비율의 개념이 성립하는 값 |


### 8. 데이터 시각화
#### 데이터 시각화의 정의
- 데이터를 쉽게 이해할 수 있도록 시각적으로 표현하고 전달하는 과정
- 데이터 사이의 관계를 대표할 수 있는 특징을 도표나 차트로 표현

slide15-image1


#### 선 차트(Line Chart)
- 목적: 추세(Trend), 패턴, 변동성을 파악
- X 축: 순서(시간, 위치, 단계), 범주형
- Y 축: 수치형
- 선 색상: 범주형
- 선 모양: 범주형 변수
- 추세선을 통해 변화 추이를 파악 가능

slide16-image1


- 추세선: 현재 기준 점에서의 전후 데이터의 평균을 출력. Ex) 이동 평균

slide17-image1


#### 상호작용 차트: 두 범주형 변수의 대상 변수에 대한 상호 관련성 출력
- X 축: 순서(시간, 위치, 단계), 범주형
- Y 축: 수치형
- 선 색상: 범주형

slide18-image1


#### 막대 차트(Bar Chart)

- 목적: 값 비교, 범주의 분포 형태 관찰
- X 축: 범주, 시간, 순서
- Y 축: 수치형 데이터
- 막대(Bar): 값의 크기
- 막대 색상(Bar Color): 범주형
- 추세선을 통해 변화 패턴을 파악 가능

#### 파레토 차트
- 파레토 법칙: 결과의 80%가 원인의 20%에서 나온다.
- 막대 그래프: 객체별 출현 빈도를 내림 차수
- 선 그래프: 객체의 누적 점유율을 그래프로 표시

slide20-image1


#### 히스토그램(Histogram)
- 목적: 데이터의 분포 파악
- X 축: 범위를 나눈 구간
- Y 축: 구간의 빈도
- 붙어 있는 막대: 빈도
- 막대 색상(Bar Color): 범주형

slide21-image1


#### 커널 밀토 플롯(Kernel Density Plot, KDE plot)

- 목적: 데이터의 분포 파악
- X 축: 연속형 변수
- Y 축: 추정된 밀도값
- 곡선의 색상: 범주형

slide22-image1


#### 산점도(Scatter plot)
- 목적: 두 변수 간의 패턴, 상관 관계, 분포, 이상치
- X 축: 수치형
- Y 축: 수치형
- 점의 크기: 수치형
- 점의 색상: 범주형
- 점의 모양: 범주형
- 점의 진함: 밀도
- 추세선을 통해 변화 패턴을 파악 가능

- 유사할 수록 직선

#### 파이 차트(Pie Chart)

- 목적: 범주형 변수의 분포, 구성비 비교
- 레이블과 색상: 범주
- 조각: 범주값의 비율

slide26-image1


#### 박스 플롯(Box Plot)

- 목적: 분포 모양, 중앙값, 이상치 여부 파악 <br/> 여러 그룹 간의 분포 비교
- - Q1(제 1 사분위수), Q2(중앙값), Q3(제 3 사분위수)
- 박스: IQR
- 하한선(Lower Whisker): Q1 – 1.5×IQR
- 상한선(Upper Whisker): Q3 – 1.5×IQR
- 점: 이상치(Outlier)

- X 축: 범주형
- Y 축: 범주형

#### 등고선 그래프

#### 바이올린 플롯(Violin Plot)
- 날개:  연속형 변수의 분포
- X 축 범주형 변수를 축에 배치하여 범주형 변수와 연속형 변수의 상관관계 시각화
- Y 축: 연속형 변수와 4 분위 통계
- 좌우측 구분: 바이올린의 축을 기준으로 이진변수에 따른 구분 출력

#### 줄기와 잎 플롯(Stem & Leaf 플롯)

- 줄기(Stem) / 잎(Leaf) 파트로 구성
- 줄기: 10 이상의 자리수
- 잎: 해당 줄기의 1 자리 숫자의 빈도수
- 빈도의 효율적 출력
- 줄기에 해당하는 빈도와 줄기 내에서의 1 자리 단위의 빈도수까지 확인 가능
