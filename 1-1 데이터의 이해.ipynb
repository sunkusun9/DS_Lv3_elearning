#### 빅데이터 특징

| 광의 | 협의 | 정의 | 특징 |
| --- | --- | --- | --- |
| 5V | 3V | Volume(규모) | 데이터 양의 급격한 증가<br/>→ 전통적 데이터 시스템으로 처리에 한계 |
|   |   | Variety(유형) | 데이터 종류와 소스의 다양화<br/>→ 정형 데이터 뿐만 아니라 반정형, 비정형 데이터로 확대 |
|   |   | Velocity(속도) | 데이터가 생성되고 전송되는 속도가 매우 빠름<br/>→ 데이터 수집 및 처리의 고속화 |
|   | 2V | Veracity(품질) | 데이터의 신뢰성, 정확성, 타당성 보장이 필수 |
|   |   | Value(가치) | 대용량 데이터에서는 가치는 내재되어 있음 방대한 데이터에서 인사이트를 도출할 수 있는 역량이 필요 |

**[Ex.4]** 

<div style="border: 1px solid #ddd; padding: 12px; margin-top: 10px;">

df_space에서 Cabin 변수는 다음과 같이 쪼갤 수 있습니다.

Cabin: The cabin number where the passenger is staying. Takes the form deck/num/side, where side can be either P for Port or S for Starboard.

Cabin에 따른 Transported 성공률을 선 차트를 이용하여 알아 봅니다.
    
</div>

````python
f, axes = plt.subplots(2, 1, figsize=(8, 6))
df = df_space['Cabin'].str.split('/', expand=True).rename(columns ={0: 'CabinDeck', 1: 'CabinNum', 2: 'CabinSide'}).assign(
    Transported = df_space['Transported']
).sort_values('CabinNum').groupby(['CabinDeck', 'CabinSide']).rolling(50, min_period=10)['Transported'].mean().dropna().reset_index().rename(
    columns={'level_2': 'CabinNum'}
)
for side, ax in zip(df['CabinSide'].unique(), axes):
    sns.lineplot(
        df.loc[df['CabinSide'] == side], x = 'CabinNum', y = 'Transported', hue = 'CabinDeck', ax=ax
    )
    ax.set_title('Cabin side: {}'.format(side))
plt.tight_layout()
plt.show()
````

<div style="border: 1px solid #ddd; padding: 12px; margin-top: 10px;">

df_space에서 연령대(Age 기준으로 10살 간격으로 구분)에 따른 Transported 성공율을 막대 차트로 출력합니다.

````python
plt.figure(figsize = (8, 3))
sns.barplot(
    df_space.groupby(pd.cut(df_space['Age'], np.arange(0, 80, 10).tolist() + [np.inf]))['Transported'].mean().reset_index(),
    y = 'Transported', x= 'Age'
)
plt.gca().set_xticklabels(plt.gca().get_xticklabels(), rotation=90)
plt.show()
````

#### 파레토 차트

**[Ex.16]** 

<div style="border: 1px solid #ddd; padding: 12px; margin-top: 10px;">

온도와 습도에 따른 이슬점의 관계를 등고선 그래프로 출력합니다.

</div>

````python
# 이슬점 계산 함수
def calculate_dew_point(temp, rh):
    a = 17.62
    b = 243.12
    gamma = (a * temp) / (b + temp) + np.log(rh / 100.0)
    dew_point = (b * gamma) / (a - gamma)
    return dew_point

# 기온과 습도 범위 설정
temp_range = np.linspace(-10, 40, 100)  # -10°C to 40°C
rh_range = np.linspace(10, 100, 100)    # 10% to 100%

T, RH = np.meshgrid(temp_range, rh_range)
DewPoint = calculate_dew_point(T, RH)

# 등고선 그래프 그리기
plt.figure(figsize=(10, 6))
contour = plt.contourf(T, RH, DewPoint, cmap='coolwarm', levels=20)
cbar = plt.colorbar(contour)
cbar.set_label('Dew Point (°C)')

plt.title('Dew Point by Temperature and Humidity')
plt.xlabel('Temperature (°C)')
plt.ylabel('Relative Humidity (%)')
plt.grid(True)
plt.show()
````
